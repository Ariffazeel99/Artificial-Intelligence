```python
import pandas as pd
dataset = pd.read_csv('cancer.csv')
```

```python
x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
y = dataset["diagnosis(1=m, 0=b)"]
```

```python
from sklearn.model_selection import train_test_split
# split training data and testing data.

x_training, x_testing, y_training, y_testing = train_test_split(x,y, test_size=0.3)
# 20 perscent data for evaluation


```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(256,input_shape=x_training.shape, activation='sigmoid'))
# 256 bubbles and sigmoid is basically values from 0 to 1 f(x)= 1/1+e-Bx
#model.add(tf.keras.layers.Dense(256, activatation='sigmoid')) #
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))# 0- 1 thats why 1 here
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Now Train it on our data
model.fit(x_training, y_training, epochs=500)
```

```python
# now final thing the evaluation
model.evaluate(x_testing,y_testing)
```
